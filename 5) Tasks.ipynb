```python
import numpy as np
import pandas as pd
```

# Creating a bank dataset with shape(5,5)

```python
a1={"bank":["ICICI","ICICI","SBI","BOB","YES BANK"],"acc handlers":[4,4,2,np.nan,1],"acc_type":["savings","savings",np.nan,"credit","credit"],"grp_id":[109,103,102,106,107],"int rate":[5,5,6,7,8]}
a1
```

```python
a2=pd.DataFrame(a1)  #converting is to dataframe
a2
```

# Changing records in 2 columns

```python
a2["acc handlers"][2]=8.0  #2.0 changed to 8.0
a2["grp_id"][1]=109        #103 changed to 109
a2
```

# Delete 1 column

```python
del a2["int rate"]
a2
```

# Detect and treat missing values

```python
a2.isnull()
```

```python
a2.isna().sum()
```

```python
a2["acc handlers"].mean()
```

```python
a2["acc handlers"].median()
```

```python
a2["acc handlers"].fillna(4.0, inplace = True)
a2
```

```python
a2["acc_type"].fillna("debit", inplace = True)
a2
```
